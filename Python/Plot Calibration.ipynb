```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#import calibration files skipping header
cal1 = pd.read_csv('/home/gsilvest/Work/miniTRB data/HF000T/draft/HF000T_PA_008.cal', skiprows=18, header= None)
#cal1 = pd.read_csv('/home/gsilvest/Work/miniTRB data/PPTM000T/draft/PPTM000T_PA_000.cal', skiprows=18, header= None)
```

```python
#plotting pedestal
cal1.plot(kind='line',x=0,y=3,color='red')
plt.show()
```

```python
%ls
```

```python
ped = cal1[3].to_numpy()
sigma_raw = cal1[4].to_numpy()
sigma = cal1[5].to_numpy()
```

```python
#plotting

fig=plt.figure(figsize=(16, 9), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(ped, '--bx')
major_ticks = np.arange(0, 640, 64)
plt.title('Pedestals')
plt.xlabel('Channel')
plt.ylabel('ADC')
axes = plt.gca()
axes.set_xticks(major_ticks)
axes.set_xlim([0,640])
axes.grid(which='both')
```

```python
fig=plt.figure(figsize=(16, 9), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(sigma_raw, '--rx')
major_ticks = np.arange(0, 640, 64)
plt.title('Sigma Raw')
plt.xlabel('Channel')
plt.ylabel('ADC')
axes = plt.gca()
axes.set_xticks(major_ticks)
axes.set_xlim([0,640])
axes.grid(which='both')
```

```python
fig=plt.figure(figsize=(16, 9), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(sigma, '--yx')
major_ticks = np.arange(0, 640, 64)
plt.grid()
plt.title('Sigma')
plt.xlabel('Channel')
plt.ylabel('ADC')
axes = plt.gca()
axes.set_xticks(major_ticks)
axes.set_xlim([0,640])
```
